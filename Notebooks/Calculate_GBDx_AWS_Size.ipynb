```python
allRes = []
with open(spFile, 'r') as inFile:
    for line in inFile:
        fileInfo = line[31:]
        fileSizeSplit = line[:31].split(" ")
        try:
            fileSize = float(fileSizeSplit[-1])
        except:
            fileSize = float(fileSizeSplit[-2])
        """splitFolder = line.split(" ")
        cFile = splitFolder[-1].replace("\n", "")
        cSize = float(splitFolder[-2])
        """
        allRes.append([os.path.dirname(fileInfo), os.path.basename(fileInfo), fileSize])
        
```

```python
res = pd.DataFrame(allRes, columns=['filePath','fileName', 'size'])
```

```python
depth = 3
res['curPath'] = res['filePath'].apply(lambda x: "_".join(x.split("/")[:depth]))
fileSize = res.groupby(['curPath'])['size'].sum()/1000000000
xx = pd.DataFrame(fileSize).reset_index()
xx.sort_values(by=['size'],ascending =False)
```
